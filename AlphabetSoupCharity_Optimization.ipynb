```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns = ['EIN', 'NAME'])
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_counts = application_df["APPLICATION_TYPE"].value_counts()
application_counts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
# cutoff value 737
application_types_to_replace = list(application_counts.iloc[6:].index)

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"OtherApplication")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
pd.set_option('display.max_rows', None)
classification_counts = application_df["CLASSIFICATION"].value_counts()
classification_counts
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classification_value_counts = application_df["CLASSIFICATION"].value_counts()
classification_greater_1 = classification_value_counts[classification_value_counts > 1]
print(classification_greater_1)
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
# cutoff value 777
classifications_to_replace = list(classification_counts.iloc[6:].index)

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"OtherClassification")

# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Renaming the "other" values in the AFFILIATION and USE_CASE columns
other_affiliation = {'Other': 'OtherAffiliation'}
application_df['AFFILIATION'] = application_df['AFFILIATION'].replace(other_affiliation)

other_use_case= {'Other': 'OtherUseCase'}
application_df['USE_CASE'] = application_df['USE_CASE'].replace(other_use_case)
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
dummies_df = pd.get_dummies(application_df)
dummies_df.head()
```

```python
# Making sure dtypes are int
dummies_df.dtypes
```

```python
# Getting a list of the column names
column_names = dummies_df.columns.tolist()
print(column_names)
```

```python
# Removing the label column
column_names.remove("IS_SUCCESSFUL")
print(column_names)
```

```python
# Split our preprocessed data into our features and target arrays

# Separting out the y variable, the labels
y=application_df["IS_SUCCESSFUL"]
# Separting out the X variable, the features
# Including all of the possibile features in the first model (everthing except IS_SUCCESSFUL)
X=dummies_df[column_names]

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train_scaled[0])
hidden_nodes_layer1 =  27
hidden_nodes_layer2 = 25
# Adding an additional hidden layer
hidden_nodes_layer3 = 25

nn_2 = tf.keras.models.Sequential()

# First hidden layer
nn_2.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1,
    input_dim=number_input_features, activation="relu")
)

# Second hidden layer
nn_2.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Third hidden layer
nn_2.add(tf.keras.layers.Dense(units=hidden_nodes_layer3, activation="relu"))

# Output layer
nn_2.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn_2.summary()
```

```python
# Compile the model
nn_2.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn_2.fit(X_train_scaled,y_train,epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn_2.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
nn_2.save("AlphabetSoupCharity_2.h5")
```
