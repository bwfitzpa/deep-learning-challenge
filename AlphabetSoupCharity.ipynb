## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns = ['EIN', 'NAME'])
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_df["APPLICATION_TYPE"].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace = ['T4','T6','T5','T19','T8','T7','T10','T9','T13','T12','T2',
                'T25','T14','T29','T15','T17']

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"OtherApplication")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
pd.set_option('display.max_rows', None)
application_df["CLASSIFICATION"].value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classification_value_counts = application_df["CLASSIFICATION"].value_counts()
classification_greater_1 = classification_value_counts[classification_value_counts > 1]
print(classification_greater_1)
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = ['C2000','C1200','C3000','C2100','C7000','C1700','C4000',
                              'C5000','C1270','C2700','C2800','C7100','C1300','C1280','C1230',
                              'C1400','C7200','C2300','C1240','C8000','C7120','C1500','C1800',
                              'C6000','C1250','C8200','C1238','C1278','C1235','C1237','C7210',
                              'C2400','C1720','C4100','C1257','C1600','C1260','C2710','C0',
                              'C3200','C1234','C1246','C1267','C1256','C2190','C4200','C2600',
                              'C5200','C1370','C1248','C6100','C1820','C1900','C1236','C3700',
                              'C2570','C1580','C1245','C2500','C1570','C1283','C2380','C1732',
                              'C1728','C2170','C4120','C8210','C2561','C4500','C2150']

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"OtherClassification")

# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Renaming the "other" values in the AFFILIATION and USE_CASE columns
other_affiliation = {'Other': 'OtherAffiliation'}
application_df['AFFILIATION'] = application_df['AFFILIATION'].replace(other_affiliation)

other_use_case= {'Other': 'OtherUseCase'}
application_df['USE_CASE'] = application_df['USE_CASE'].replace(other_use_case)
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_type_dummies = pd.get_dummies(application_df["APPLICATION_TYPE"]).astype(int)
affiliation_categories = pd.get_dummies(application_df["AFFILIATION"]).astype(int)
classification_dummies = pd.get_dummies(application_df["CLASSIFICATION"]).astype(int)
use_case_categories = pd.get_dummies(application_df["USE_CASE"]).astype(int)
organization_categories = pd.get_dummies(application_df["ORGANIZATION"]).astype(int)
income_amt_categories = pd.get_dummies(application_df["INCOME_AMT"]).astype(int)
special_considerations_dummies = pd.get_dummies(application_df["SPECIAL_CONSIDERATIONS"]).astype(int)
application_df = pd.concat([application_df, application_type_dummies, affiliation_categories,
                            classification_dummies, use_case_categories, organization_categories,
                            income_amt_categories, special_considerations_dummies], axis=1)
application_df.head()
```

```python
# Getting a list of the column names
column_names = application_df.columns.tolist()
print(column_names)
```

```python
# Split our preprocessed data into our features and target arrays

# Separting out the y variable, the labels
y=application_df["IS_SUCCESSFUL"]
# Separting out the X variable, the features
# Including all of the possibile features for now
X=pd.DataFrame(application_df, columns=[ 'OtherApplication', 'T3', 'CompanySponsored', 'Family/Parent',
                                        'Independent', 'National', 'OtherAffiliation', 'Regional',
                                         'C1000', 'OtherClassification', 'CommunityServ', 'Heathcare',
                                         'OtherUseCase', 'Preservation', 'ProductDev', 'Association',
                                         'Co-operative', 'Corporation', 'Trust', '0', '1-9999', '10000-24999',
                                         '100000-499999', '10M-50M', '1M-5M', '25000-99999', '50M+', '5M-10M',
                                         'N', 'Y'])

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train_scaled[0])
hidden_nodes_layer1 =  10
hidden_nodes_layer2 = 5

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1,
    input_dim=number_input_features, activation="relu")
)

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled,y_train,epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
nn.save("AlphabetSoupCharity.h5")
```
